visualisadi data haberman

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (upload file dulu di Colab)
df = pd.read_csv('/content/haberman.csv')

# Tambahkan nama kolom
df.columns = ['age', 'year', 'nodes', 'status']

df.head()
```

```python
print(df.info())
print(df.describe())
print(df['status'].value_counts())
```

```python
df.hist(figsize=(10,6))
plt.suptitle("Histogram Dataset Haberman")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot Dataset Haberman")
plt.show()
```

```python
sns.countplot(x='status', data=df)
plt.title("Distribusi Status Survival")
plt.show()
```

```python
sns.pairplot(df, hue='status')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x='age', y='nodes', hue='status', data=df)
plt.title("Age vs Nodes")
plt.show()
```

```python
sns.violinplot(x='status', y='age', data=df)
plt.title("Violin Plot Age vs Status")
plt.show()
```
